```python
df.groupby('account_no').Amount.sum().reset_index()
```

```python
df['period'] = df.Date.apply(lambda x: (x.year, x.month))
trial_balance = df.groupby(['period', 'account_no']) \
    .Amount.sum().reset_index() \
    .pivot(index='account_no', columns='period', values='Amount') \
    .fillna(0)
trial_balance
```

```python
house = trial_balance.loc[500].reset_index()
house
```

```python
from matplotlib import pyplot as plt
x = house.period.apply(lambda x: f'{x[1]}/{x[0]}')
y = house[500]
plt.plot(x, y)
plt.title('Account #500 Balance by Period')
```
